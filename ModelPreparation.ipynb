# Used cars price Prediction
### In this ML Project i have predicted sales values for used cars.I have used RandomForestRegressor algorithm for this prediction

### Project Prerequisites

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Preparing the DataFrame

```python
train_df = pd.read_csv("Data/train-data.csv")
train_df.head()
```

```python
train_df2 = train_df.drop(columns=["Unnamed: 0","Name","Location","New_Price","Fuel_Type","Power","Engine","Mileage"]) 
train_df2.head()
```

### Changing the Owner_Type String column to Integer type so that we can use it in the prediction for more accuracy 

```python
train_df2.groupby(by="Owner_Type").describe()
```

```python
train_df2['Owner_Type'] = train_df2['Owner_Type'].replace(to_replace='First', value=1)
train_df2['Owner_Type'] = train_df2['Owner_Type'].replace(to_replace='Second', value=2)
train_df2['Owner_Type'] = train_df2['Owner_Type'].replace(to_replace='Third', value=3)
train_df2['Owner_Type'] = train_df2['Owner_Type'].replace(to_replace='Fourth & Above', value=4)
train_df2.head()
```

```python
train_df2.groupby(by="Owner_Type").describe()
```

### Changing the Transmission String column to Integer type so that we can use it in the prediction for more accuracy 

```python
train_df2.groupby(by="Transmission").describe()
# so lets assain 1 for Automatic and 2 for Mannual in the data set
```

```python
train_df2["Transmission"] = train_df2["Transmission"].replace(to_replace='Automatic', value=1)
train_df2["Transmission"] = train_df2["Transmission"].replace(to_replace='Manual', value=2)
train_df2.head()
```

### Handling NaN's

```python
train_df2.isnull().sum()
```

```python
train_df2["Seats"].fillna(train_df2["Seats"].mean(),inplace=True)
```

```python
train_df2.isnull().sum()
```

### Model Validation

```python
car_features = ["Year","Kilometers_Driven","Transmission","Owner_Type","Seats"]
X = train_df2[car_features]
```

```python
y = train_df2.Price
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
model.fit(X,y)
```
