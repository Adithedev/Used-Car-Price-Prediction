# Used cars price Prediction
### In this ML Project i have predicted sales values for used cars.I have used the algorithm which we saved earlier in the notebook

### Project Prerequisites

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Preparing the DataFrame

```python
test_df = pd.read_csv("Data/test-data.csv")
test_df.head()
```

```python
test_df2 = test_df.drop(columns=["Unnamed: 0","Name","Location","New_Price","Fuel_Type","Power","Engine","Mileage"]) 
test_df2.head()
```

```python
test_df2['Owner_Type'] = test_df2['Owner_Type'].replace(to_replace='First', value=1)
test_df2['Owner_Type'] = test_df2['Owner_Type'].replace(to_replace='Second', value=2)
test_df2['Owner_Type'] = test_df2['Owner_Type'].replace(to_replace='Third', value=3)
test_df2['Owner_Type'] = test_df2['Owner_Type'].replace(to_replace='Fourth & Above', value=4)
test_df2.head()
```

```python
test_df2["Transmission"] = test_df2["Transmission"].replace(to_replace='Automatic', value=1)
test_df2["Transmission"] = test_df2["Transmission"].replace(to_replace='Manual', value=2)
test_df2.head()
```

```python
test_df2["Seats"].fillna(test_df2["Seats"].mean(),inplace=True)
test_df2.isnull().sum()
```

### Predicting the Price

```python
import joblib
model = joblib.load("used_car_prediction.joblib")
```

```python
predictions = model.predict(test_df2)
```

### Saving the Predicted Price in a new DataFrame

```python
predicted_df = pd.DataFrame(data = predictions)
predicted_df.head()
```

```python
predicted_df.rename(columns = {0:"Price"},inplace = True)
```

```python
predicted_df["Year"] = test_df2["Year"]
predicted_df["Kilometers_Driven"] = test_df2["Kilometers_Driven"]
predicted_df["Transmission"] = test_df2["Transmission"]
predicted_df["Owner_Type"] = test_df2["Owner_Type"]
predicted_df["Seats"] = test_df2["Seats"]
predicted_df.head()
```

```python
predicted_df.to_csv("used_car_predictions.csv")
```
